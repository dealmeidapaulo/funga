```python
import geopandas as gpd
import pandas as pd

```

```python
# Read the GeoDataFrame from file
gdfPolygons = gpd.read_file('polygons/polygons.shp')

# Export as GeoJSON
output_file = 'polygons/polygons.geojson'
gdfPolygons.to_file(output_file, driver='GeoJSON')
```

```python
# Read the GeoDataFrame from file
gdfObservations = gpd.read_file('observations/observations.shp')

# Export as GeoJSON
output_file = 'observations/observations.geojson'
gdfObservations.to_file(output_file, driver='GeoJSON')
```
