```python
import geopandas as gpd
from shapely.geometry import Point 


import json

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to the CSV file
gdf_path = 'polygons/pxdptodatosok.shp'

# Read the CSV file into a Geopandas DataFrame
gdfPolygons = gpd.read_file(gdf_path)
gdfPolygons['#obs'] = np.nan



gdfPolygons.to_file(gdf_path, index=False)
```

```python
# Path to the CSV file
csv_path = 'observations/observationsArgentinaRaw.csv'

# Read the CSV file into a Geopandas DataFrame
gdfObservations = gpd.read_file(csv_path)

# Create a new geometry column with Point objects
geometryColumn = [Point(xy) for xy in zip(gdfObservations['longitude'], gdfObservations['latitude'])]

# Assign the geometry column to the Geopandas DataFrame
gdfObservations['geometry'] = geometryColumn

# Drop the latitude and longitude columns
gdfObservations = gdfObservations.drop(['latitude', 'longitude'], axis=1)

# Create a new  column polygon
gdfObservations['polygon'] = np.nan

# Save the modified Geopandas DataFrame as a new CSV file
gdfObservations.to_file('observations/observationsArgentina.shp')
```
